```python
import pandas as pd
import nltk
import re
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from wordcloud import WordCloud
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
import xgboost as xgb
import torch
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder  




# Load the dataset
file_path = r"C:\Users\ajayg\Downloads\sample_data.csv"
data = pd.read_csv(file_path)

# Check the structure of the dataset
print("\nDataset Structure:")
print(data.info())

# Summary statistics for any numerical columns
print("\nSummary Statistics:")
print(data.describe())

# Explore the categories in the 'label' column
print("\nCategory Distribution:")
print(data['label'].value_counts())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Visualize the distribution of categories
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='label')
plt.title('Distribution of Categories')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

import nltk

# Download the German tokenizer resource
nltk.download('punkt', download_dir='/usr/local/share/nltk_data')

data_cleaned = data.dropna(subset=['label'])
# Check for duplicates in the 'text' column
duplicates = data_cleaned[data_cleaned.duplicated(subset=['text'], keep='first')]

# Remove duplicates (keep the first occurrence)
data_cleaned = data_cleaned.drop_duplicates(subset=['text'], keep='first')

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import spacy

nltk.download('stopwords')
stop_words = set(stopwords.words('german'))

# Modified preprocess_text function
def preprocess_text(text, use_stemming=True, use_lemmatization=False):
    # Tokenize text using NLTK
    tokens = nltk.word_tokenize(text, language='german')
    
    # Lowercase and filter out non-alphabetic tokens and stopwords
    tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    
    # Apply stemming
    if use_stemming:
        stemmer = SnowballStemmer('german')
        tokens = [stemmer.stem(word) for word in tokens]
    
    # Apply lemmatization
    if use_lemmatization:
        nlp = spacy.load('de_core_news_sm')
        doc = nlp(' '.join(tokens))
        tokens = [token.lemma_ for token in doc]

    return ' '.join(tokens)

# Apply text preprocessing to the 'text' column
data_cleaned['text'] = data_cleaned['text'].apply(preprocess_text)

# Word Cloud for Common Words
category_wordclouds = {}
categories = data_cleaned['label'].unique()

for category in categories:
    category_text = " ".join(data_cleaned[data_cleaned['label'] == category]['text'])
    wordcloud = WordCloud(width=800, height=400).generate(category_text)
    category_wordclouds[category] = wordcloud

# Plot word clouds for each category
plt.figure(figsize=(12, 8))
for i, (category, wordcloud) in enumerate(category_wordclouds.items(), 1):
    plt.subplot(2, 3, i)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for {category}')
    plt.axis("off")
plt.tight_layout()
plt.show()

# Bar Chart for Top N Words
top_n = 10
top_words = pd.Series(" ".join(data_cleaned['text']).split()).value_counts()[:top_n]
plt.figure(figsize=(10, 6))
sns.barplot(x=top_words.values, y=top_words.index, palette="viridis")
plt.title(f'Top {top_n} Most Common Words')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()

# Histogram for Text Length
data_cleaned['text_length'] = data_cleaned['text'].apply(lambda x: len(x.split()))
plt.figure(figsize=(8, 6))
sns.histplot(data=data_cleaned, x='text_length', bins=30, kde=True)
plt.title('Distribution of Text Length')
plt.xlabel('Text Length (Number of Words)')
plt.ylabel('Count')
plt.show()

# Box Plot for Text Length by Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_cleaned, x='label', y='text_length')
plt.title('Text Length by Category')
plt.xlabel('Categories')
plt.ylabel('Text Length (Number of Words)')
plt.xticks(rotation=45)
plt.show()

# Split the data into training (70%), validation (15%), and test (15%) sets
train_data, temp_data = train_test_split(data_cleaned, test_size=0.3, random_state=42)
valid_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# Create a LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder on the unique labels in your dataset
label_encoder.fit(train_data['label'])

# Transform the labels into integers
train_data['label_encoded'] = label_encoder.transform(train_data['label'])
valid_data['label_encoded'] = label_encoder.transform(valid_data['label'])
test_data['label_encoded'] = label_encoder.transform(test_data['label'])

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Fit and transform the training data
X_train = tfidf_vectorizer.fit_transform(train_data['text'])

# Transform the validation and test data
X_valid = tfidf_vectorizer.transform(valid_data['text'])
X_test = tfidf_vectorizer.transform(test_data['text'])

# Create a list of models to evaluate
models = [
    ("Random Forest", RandomForestClassifier(random_state=42)),
    ("Logistic Regression", LogisticRegression(random_state=42, max_iter=1000)),
    ("Support Vector Machine", SVC(random_state=42)),
    ("Multinomial Naive Bayes", MultinomialNB()),
    ("K-Nearest Neighbors", KNeighborsClassifier(n_neighbors=5)),
    ("Gradient Boosting", GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ("XGBoost", xgb.XGBClassifier(random_state=42)),
    ("Neural Network", MLPClassifier(random_state=42)),
]

# Evaluate models
results = []
for name, model in models:
    model.fit(X_train, train_data['label_encoded'])  # Train with encoded labels
    valid_predictions = model.predict(X_valid)
    accuracy = accuracy_score(valid_data['label_encoded'], valid_predictions)  # Compare with encoded labels
    f1 = f1_score(valid_data['label_encoded'], valid_predictions, average='weighted')  # Use encoded labels
    results.append((name, accuracy, f1))


# Display results
results_df = pd.DataFrame(results, columns=['Model', 'Accuracy', 'F1 Score'])
print(results_df)

```

```python
from sklearn.metrics import confusion_matrix

# Plot confusion matrix for the best-performing model (Random Forest)
best_model = RandomForestClassifier(random_state=42)
best_model.fit(X_train, train_data['label_encoded'])
test_predictions = best_model.predict(X_test)

# Calculate confusion matrix
conf_matrix = confusion_matrix(test_data['label_encoded'], test_predictions)

# Plot confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(test_data['label_encoded'], test_predictions, display_labels=categories, cmap="Blues", xticks_rotation='vertical')
plt.title('Confusion Matrix for Random Forest')
plt.show()



# BERT-based Language Model for Text Classification
# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-german-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=len(categories))

# Tokenize and encode the data
train_encodings = tokenizer(list(train_data['text']), truncation=True, padding=True, return_tensors='pt', max_length=256)
valid_encodings = tokenizer(list(valid_data['text']), truncation=True, padding=True, return_tensors='pt', max_length=256)
test_encodings = tokenizer(list(test_data['text']), truncation=True, padding=True, return_tensors='pt', max_length=256)

# Create PyTorch datasets and dataloaders
train_dataset = TensorDataset(train_encodings['input_ids'], train_encodings['attention_mask'], torch.tensor(list(train_data['label_encoded'])))
valid_dataset = TensorDataset(valid_encodings['input_ids'], valid_encodings['attention_mask'], torch.tensor(list(valid_data['label_encoded'])))
test_dataset = TensorDataset(test_encodings['input_ids'], test_encodings['attention_mask'], torch.tensor(list(test_data['label_encoded'])))

batch_size = 32
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_dataloader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Set up training parameters
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)
epochs = 3

# Training loop
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for batch in tqdm(train_dataloader, desc=f"Epoch {epoch + 1}"):
        input_ids, attention_mask, labels = batch
        input_ids, attention_mask, labels = input_ids.to(device), attention_mask.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    average_loss = total_loss / len(train_dataloader)
    print(f"Epoch {epoch + 1}: Average Loss: {average_loss:.4f}")

# Evaluation
model.eval()
predictions = []
true_labels = []

for batch in tqdm(valid_dataloader, desc="Validation"):
    input_ids, attention_mask, labels = batch
    input_ids, attention_mask, labels = input_ids.to(device), attention_mask.to(device), labels.to(device)
    
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
    
    logits = outputs.logits
    predictions.extend(torch.argmax(logits, dim=1).tolist())
    true_labels.extend(labels.tolist())

# Calculate accuracy and F1 score for BERT
bert_accuracy = accuracy_score(true_labels, predictions)
bert_f1 = f1_score(true_labels, predictions, average='weighted')

print(f"BERT Validation Accuracy: {bert_accuracy}")
print(f"BERT F1 Score: {bert_f1}")

# Confusion Matrix for BERT Model
conf_matrix = confusion_matrix(test_data['label_encoded'], test_predictions)

# Plot confusion matrix for BERT model
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(true_labels, predictions, display_labels=categories, cmap="Blues", xticks_rotation='vertical')
plt.title('BERT Confusion Matrix')
plt.show()

```
